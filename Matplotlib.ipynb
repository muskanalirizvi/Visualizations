```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('laptop_price.csv', encoding = 'ISO-8859-1')
df.head()
```

```python
plt.bar(df['OpSys'], df['Price_euros'], color = 'red')
plt.xlabel('Operating Systems', )
plt.ylabel('Price in euros')
plt.xticks(rotation = 90)
plt.title('Laptop Prices by Operating System', color = 'blue')

plt.show()
```

```python
plt.scatter(df['Inches'], df['Price_euros'], c = 'red',  marker = 'o')
plt.xticks(rotation = 90)
plt.xlabel('Size in inches')
plt.ylabel('Price in euros')
plt.title('Price vs Screen Size (Inches)', color = 'blue')
plt.show()
```

```python
df['TypeName'].value_counts()
```

```python
count = [727, 205 ,196, 121, 29, 25]
labels = ['Notebook', 'Gaming', 'Ultrabook', '2 in 1 Convertible', 'Workstation', 'Netbook']

plt.pie(count, labels=labels, autopct='%1.1f%%', startangle=0)
plt.axis('equal')
plt.title('Distribution of Laptop Types', pad=40, color = 'blue')
plt.show()
```

```python
# Histogram for screen size
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)  # 2 rows, 3 columns, first subplot
plt.hist(df['Inches'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Screen Sizes')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Frequency')

# Histogram for RAM
plt.subplot(2, 3, 2)
plt.hist(df['Ram'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Frequency')
plt.xticks(rotation = 90)

# Histogram for Price
plt.subplot(2, 3, 3)
plt.hist(df['Price_euros'], bins= 7, color='salmon', edgecolor='black')
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price (Euros)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
